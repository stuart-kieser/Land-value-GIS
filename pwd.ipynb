The following document showcases the use of data analysis modules such as numpy, and matplotlib to showcase the distibution of of house prices in South Africa.

```python
import matplotlib.pyplot as plt 
import numpy as np
import os
import csv
```

Numpy and matplotlib are modules commonly used in data analysis of big data. Often being more effective than the like of standard python these two modules are written in the C programing language.

```python
target_directory = r'C:\GitHub\Land-value-GIS\backups'

locations_price_paths = [
    "prop_westerncape_locations_price.csv",

]

"""    "prop_eastern_cape_locations_price.csv",
    "prop_freestate_locations_price.csv",
    "prop_gauteng_locations_price.csv",
    "prop_kwazulu_natal_locations_price.csv",
    "prop_limpopo_locations_price.csv",
    "prop_mpumalanga_locations_price.csv",
    "prop_northern_cape_locations_price.csv",
    "prop_northwest_locations_price.csv"""

pricearray = []


for loc in locations_price_paths:
    path = os.path.join(target_directory, loc)
    with open(path, "r") as r:
        reader = csv.reader(r)
        for data in reader:
            if len(data) > 1:
                pricearray.append(data[0])
```

```python

plt.hist(pricearray, bins=6, edgecolor="pink")

# Labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with Custom Bin Ranges')
```
