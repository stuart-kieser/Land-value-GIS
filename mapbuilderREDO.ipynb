This is a demonstration of a land value GIS

```python
import numpy as np
import csv
import os

coords_directory = r'C:\GitHub\Land-value-GIS\backups'

coords_list_paths = [
    "westerncape_area_coords.csv",
    "eastern_cape_area_coords.csv",
    "freestate_area_coords.csv",
    "gauteng_area_coords.csv",
    "kwazulu_natal_area_coords.csv",
    "limpopo_area_coords.csv",
    "mpumalanga_area_coords.csv",
    "northern_cape_area_coords.csv",
    "northwest_area_coords.csv"
]

coords_list = []

# Iterate through coords_list_paths
for path in coords_list_paths:
    full_path = os.path.join(coords_directory, path)  # Create the full path
    with open(full_path, "r") as w:
        reader = csv.reader(w)

        for row in reader:
            if len(row) > 1:
                coords_list.append(row)    
```

Co-ordinates are pulled from the csv files into a list that is then used to plot onto a map.

```python
print(coords_list)
print(len(coords_list))
```

```python
from ipyleaflet import Map, basemaps, Heatmap

center = [-26, 28]
zoom = 5

m = Map(basemap=basemaps.Esri.WorldStreetMap, center=center, zoom=zoom)

heatmap = []

for prop in coords_list:
    try:
        if prop is not None:
            hprop = prop[0], prop[1], 3
            heatmap.append(hprop)
    except TypeError as e:
        print(e)

print(heatmap)

m.add_layer(Heatmap(locations=heatmap, radius=15, blur=10))

m


```
